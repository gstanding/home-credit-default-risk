```python
import pandas as pd
import numpy as np

from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import normalize
```

```python
def train_model(train_, y_, val_x_, val_y_, test_):
    feats_ = [f_ for f_ in test_.columns if f_ not in ['SK_ID_CURR']]
    val_preds = np.zeros(val_x_.shape[0])
    sub_preds = np.zeros(test_.shape[0])
    clf = SVC(probability=True)
    bag = BaggingClassifier(
        base_estimator=clf,
        n_estimators =6,
        n_jobs=-1,
        max_samples=0.01
        )
    bag.fit(train_, y_)
    val_preds = bag.predict_proba(val_x_[feats_])[:, 1]
    sub_preds = bag.predict_proba(pd.DataFrame(test_[feats_]))[:, 1]
    print('validate AUC score %.6f'%roc_auc_score(val_y_, val_preds))
    test_['TARGET'] = sub_preds
    
    return test_[['SK_ID_CURR', 'TARGET']]    
    
```
